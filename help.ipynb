```python
from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows
import pandas as pd
from copy import copy

def cope_excels(file0, file1, output_path):
    # 使用 pandas 读取数据
    df0 = pd.read_excel(file0)
    df1 = pd.read_excel(file1)

    # 将 df1 的最后 4 列添加到 df0 中，避免列名冲突
    df1_last4 = df1.iloc[:, -4:].copy()
    df1_last4.columns = [f"{col}_新" for col in df1_last4.columns]

    # 合并数据
    df_merged = pd.concat([df0, df1_last4], axis=1)
    
    # 新增一列 '两次认定结果差异性'
    df_merged['两次认定结果差异性'] = df_merged.apply(
        lambda row: '一致' if row['算法认定结果'] == row['算法认定结果_新'] else '不一致',
        axis=1
    )

    # 加载工作簿
    wb0 = load_workbook(file0)
    ws0 = wb0.active
    wb1 = load_workbook(file1)
    ws1 = wb1.active

    # 将合并后的数据写入到 ws0
    for r_idx, row in enumerate(dataframe_to_rows(df_merged, index=False, header=True), 1):
        for c_idx, value in enumerate(row, 1):
            ws0.cell(row=r_idx, column=c_idx, value=value)

    # 复制最后4列的格式
    num_cols_df0 = df0.shape[1]
    for r_idx in range(2, ws0.max_row + 1):  # 从第二行开始，跳过标题行
        for c_idx in range(1, 5):
            source_cell = ws1.cell(row=r_idx, column=ws1.max_column - 4 + c_idx)
            target_cell = ws0.cell(row=r_idx, column=num_cols_df0 + c_idx)
            if source_cell.has_style:
                target_cell.font = copy(source_cell.font)
                target_cell.border = copy(source_cell.border)
                target_cell.fill = copy(source_cell.fill)
                target_cell.number_format = copy(source_cell.number_format)
                target_cell.protection = copy(source_cell.protection)
                target_cell.alignment = copy(source_cell.alignment)

    # 保存工作簿
    wb0.save(output_path)


cope_excels(
    "D:\Working\助管\精准资助\试点验证方案\精准资助相关文件\真实结果\回复\一致数据集两次认定\交叉审核标准集结果比对\\0.xlsx",
    "D:\Working\助管\精准资助\试点验证方案\精准资助相关文件\真实结果\回复\一致数据集两次认定\交叉审核标准集结果比对\\1.xlsx",
    "D:\Working\助管\精准资助\试点验证方案\精准资助相关文件\真实结果\回复\一致数据集两次认定\交叉审核标准集结果比对\coped.xlsx"
)
```
