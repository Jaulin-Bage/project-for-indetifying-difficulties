````python
# 将四所学校实际认定结果与算法认定结果一致的找出并合并到一起
work_dir = 'D:\Working\助管\精准资助\试点验证方案\真实结果\with_diff'

import pandas as pd
import os

all_files = []
for file_name in os.listdir(work_dir):
    if file_name.endswith('.csv'):
        all_files.append(os.path.join(work_dir, file_name))

all_files
````

````python
df = pd.DataFrame()
for file in all_files:
    temp_df = pd.read_csv(file)
    df = pd.concat([df, temp_df], ignore_index=True)

df.shape
````

````python
df[df['实际认定结果'] == df['算法认定结果']].shape
````

````python
df_same = df[df['实际认定结果'] == df['算法认定结果']]
df_same.to_csv('D:/Working/助管/精准资助/试点验证方案/真实结果/with_diff/四所学校一致结果合并.csv', index=False)
````
